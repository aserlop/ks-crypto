```python
import findspark
from ks_crypto.lib.spark_conn import create_yarn_connection

findspark.init()
spark = create_yarn_connection()
```

```python
project_name = 'bigquery-public-data'
dataset_name = 'crypto_bitcoin'
table_name = 'transactions'
full_table_name = f'{project_name}:{dataset_name}.{table_name}'

date_field = 'block_timestamp'
f_min = '2019-01-01'
f_max = '2019-01-02'
```

```python
df = spark.read \
  .format("bigquery") \
  .option("table", full_table_name) \
  .option("filter", f"CAST({date_field} AS STRING) BETWEEN '{f_min}' AND '{f_max}'") \
  .load()
```

```python
df.printSchema()
df.limit(10).show()
```
